```python
# Dependencies
import pandas as pd

# File to load
file = "Resources/purchase_data.csv"

# Store file contents into data frame
df = pd.read_csv(file)

# display first few rows
df.head()
```

## Player Count

```python
# count number of players by finding unique SNs
player_count = len(df["SN"].unique())
pd.DataFrame({"Total Players": [player_count]})
```

## Purchasing Analysis (Total)

```python
# find number of unique items
item_count = len(df["Item Name"].unique())

# calculate average price and round to nearest penny
avg_price = round(df["Price"].mean(),2)

# number of purchases (corresponds to number of rows in data frame)
purchase_count = len(df)

# calculate total revenue
total_revenue = df["Price"].sum()

# display as data frame
pd.DataFrame({"Number of Unique Items": [item_count],
             "Average Price": [f"${avg_price}"],
             "Number of Purchases": [purchase_count],
             "Total Revenue": [f"${total_revenue}"]})
```

## Gender Demographics

```python
# how many genders?
df["Gender"].unique()
```

```python
# put genders into separate data frames and get counts of unique SNs for each
male_df = df[df["Gender"] == "Male"]
males = len(male_df["SN"].unique())

female_df = df[df["Gender"] == "Female"]
females = len(female_df["SN"].unique())

other_df = df[df["Gender"] == "Other / Non-Disclosed"]
others = len(other_df["SN"].unique())

# summarize in data frame
gender_demo_df = pd.DataFrame({"Gender": ["Male", "Female", "Other / Non-Disclosed"],
             "Total Count": [males, females, others],
             "Percentage of Players": [males/player_count*100, females/player_count*100, others/player_count*100]})

# format percentage column
gender_demo_df["Percentage of Players"] = gender_demo_df["Percentage of Players"].map("{:.2f}%".format)

# set index to gender
gender_demo_df = gender_demo_df.set_index("Gender")

gender_demo_df
```

## Purchasing Analysis (Gender)

```python
# group original data by gender
gender_group_df = df.groupby(["Gender"])

# get counts for each gender as a series (any column works)
purchase_counts = gender_group_df["SN"].count()
# set name of series
purchase_counts.name = "Purchase Count"

# get average purchase price for each gender and set column name
avg_purchase_prices = gender_group_df["Price"].mean()
avg_purchase_prices.name = "Average Purchase Price"

# get total purchase value for each gender and set column name
total_purchase_values = gender_group_df["Price"].sum()
total_purchase_values.name = "Total Purchase Value"

# get average total purchase per person and set column name
avg_purchase_person = total_purchase_values/gender_demo_df["Total Count"]
avg_purchase_person.name = "Avg Total Purchase per Person"

# put above series' together into a data frame
gender_purchase_analysis = pd.concat([purchase_counts, avg_purchase_prices, 
                                      total_purchase_values, avg_purchase_person], axis=1)

# format appropriate columns to show currency
gender_purchase_analysis["Average Purchase Price"] = gender_purchase_analysis["Average Purchase Price"].map("${:.2f}".format)
gender_purchase_analysis["Total Purchase Value"] = gender_purchase_analysis["Total Purchase Value"].map("${:.2f}".format)
gender_purchase_analysis["Avg Total Purchase per Person"] = gender_purchase_analysis[
    "Avg Total Purchase per Person"].map("${:.2f}".format)

gender_purchase_analysis
```

## Top Spenders

```python
# group by SN
sn_group_df = df.groupby("SN")

# make series storing purchases per user (SN) and rename appropriately
purchase_counts = sn_group_df["Price"].count()
purchase_counts.name = "Purchase Count"

# make series with average purchase price
avg_purchase_prices = sn_group_df["Price"].mean()
avg_purchase_prices.name = "Average Purchase Price"

# make series for total purchase value per person
total_purchase_values = sn_group_df["Price"].sum()
total_purchase_values.name = "Total Purchase Value"

# make data frame by combining above series'
top_spenders = pd.concat([purchase_counts, avg_purchase_prices, total_purchase_values], axis=1)

# sort by total purchase value
top_spenders = top_spenders.sort_values("Total Purchase Value", ascending=False)

# formatting
top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map("${:.2f}".format)
top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map("${:.2f}".format)

top_spenders.head()
```

## Most Popular Items

```python
# group by item
item_group_df = df.groupby(["Item ID", "Item Name"])

# make series storing purhcase count of each item and rename column
purchase_counts = item_group_df["Price"].count()
purchase_counts.name = "Purchase Count"

# make series storing item prices, rename, and format
item_prices = item_group_df["Price"].mean()
item_prices.name = "Item Price"
item_prices = item_prices.map("${:.2f}".format)

# make series storing total purchase values, rename, and format
total_purchase_values = item_group_df["Price"].sum()
total_purchase_values.name = "Total Purchase Value"
total_purchase_values = total_purchase_values.map("${:.2f}".format)

# concat above to data frame
popular_items = pd.concat([purchase_counts, item_prices, total_purchase_values], axis=1)

popular_items.sort_values("Purchase Count", ascending=False).head()
```

## Most Profitable Items

```python
# take above data frame and sort by total purchase value instead
# however, due to some data type issues, must reformat

# make copy
profitable_items = popular_items.copy()

# replace column with unformatted numbers
profitable_items["Total Purchase Value"] = item_group_df["Price"].sum()

# now sort by "Total Purchase Value" column, then format back to currency
sorted_profitable_items = profitable_items.sort_values("Total Purchase Value", ascending=False)
sorted_profitable_items["Total Purchase Value"] = sorted_profitable_items["Total Purchase Value"].map("${:.2f}".format)

sorted_profitable_items.head()
```
